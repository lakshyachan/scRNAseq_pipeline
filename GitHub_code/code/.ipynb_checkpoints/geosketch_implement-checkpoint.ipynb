```python
import geosketch
import pandas as pd
```

```python
counts = pd.read_csv("/home/lchanem1/data-abattle4/lakshmi/cuomo_2020/zihe_1_counts.csv")
```

```python
counts.head(5)
```

```python
new_row = pd.DataFrame([counts.columns], columns = counts.columns)
new_counts = pd.concat([new_row, counts])
new_counts.reset_index(drop = True, inplace = True)
new_counts.head(5)
```

```python
final_counts = new_counts.T
final_counts.head(5)
```

```python
import numpy as np
X = final_counts.to_numpy()
print(X)
```

```python
# Compute PCs.
from fbpca import pca
U, s, Vt = pca(X, k=20)
X_dimred = U[:, :20] * s[:20]
```

```python
explained_variance = np.cumsum(s**2) / np.sum(s**2)
import matplotlib.pyplot as plt
plt.plot(explained_variance)
plt.xlabel('Number of Principal Components')
plt.ylabel('Explained Variance')
plt.title('Explained Variance vs. Number of Principal Components')
plt.grid(True)
plt.show()
```
